```python
for column in udemy.columns:
    if udemy[column].dtype != 'object' and column != 'course_id':
        plt.figure(figsize=(16,8))
        sns.histplot(data=udemy,x=column,palette="flare",)
        if np.std(udemy[column]) > 100:
            plt.yscale('log')
        else:
            plt.yscale('linear')
        plt.title(label= column.capitalize()+' distribution of courses in the dataset')
        plt.show()
        print('Mean:'+str(np.mean(udemy[column])))
        print()
```

```python
plt.figure(figsize=(16,8))
sns.barplot(x=udemy['level'].value_counts().keys(),
            y=udemy['level'].value_counts(),palette='flare')
plt.ylabel(ylabel='Count')
plt.title(label='Level of course distribution in dataset')
plt.show()
```

```python
levels = ['All Levels','Beginner Level','Intermediate Level','Expert Level']
numbers = []
for i in udemy['subject'].unique():
    tempDF = udemy[udemy['subject']==i]
    for j in levels:
        numbers.append([i,j,len(tempDF[tempDF['level']==j])])
data = {
    'Subject':[i[0] for i in numbers],
    'Level':[i[1] for i in numbers],
    'Count':[i[2] for i in numbers]
}
splitDF = pd.DataFrame(numbers,columns=['Subject','Level','Count'])

plt.figure(figsize=(16,8))
sns.barplot(data=splitDF,x='Subject',y='Count',hue='Level',palette='Spectral')
plt.title(label='Level distribution matched to each subject.')
plt.show()
```
